```python
# Flipkart Web Scraping Using Python And BeautifulSoup 

# Step 1 : Installing Importants Libraries
# pip install requests  ( To Get Fetch the HTML Content from website. its helps for  returning HTML code as String. )
# pip install bs4       (bs4 library provides a BeautifulSoup method to prettify the html content)

# importing above libraries
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
# get website link for scraping
url="https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_3_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_3_0_na_na_na&as-pos=3&as-type=TRENDING&suggestionId=mobiles&requestId=309ba22c-a02d-4112-8f62-8b7bd25443a1"
Headers= ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36','Accept-Language':'eng-US, en;q=0.5'})
```

```python
# step 2. Get the HTML

response=requests.get(url, headers=Headers)
print(response)
# print(response.content)
# print(type(response.content))
```

```python
# step 3. Parse the HTML

soup = BeautifulSoup(response.content, 'html.parser')
# print(soup.prettify())
```

```python
# Function to extract Product Title
def get_title(soup):

    try:
        # Outer Tag Object
        title = soup.find("span", attrs={"class":'VU-ZEz'})
        
        # Inner NavigatableString Object
        title_value = title.text

        # Title as a string value
        title_string = title_value.strip()

    except AttributeError:
        title_string = "#"

    return title_string

# Function to extract Product Price
def get_price(soup):

    try:
        price = soup.find("div", attrs={'class':'Nx9bqj CxhGGd'}).string.strip()

    except AttributeError:
            price = "#"

    return price

# Function to extract Product Rating
def get_rating(soup):

    try:
        rating = soup.find("div", attrs={'class':'XQDdHH'}).text.strip()
    
    except AttributeError:
            rating = "#"	

    return rating

# Function to extract Number of User Reviews
def get_review(soup):
    try:
        review = soup.find("span", attrs={'class':'Wphh3N'}).text

    except AttributeError:
        review = "#"	

    return review
```

```python
# Transfrom Data in pandas DataFrame
anchors = soup.find_all("a", attrs={'class':"CGtC98"})
product=[]
for i in anchors:
    product.append(i.get("href"))
# print(product)

dic={"title":[],"price":[],"rating":[],"review":[]}

for link in product:
    new_response= requests.get("https://www.flipkart.com"+link,headers=Headers)
    new_soup=BeautifulSoup(new_response.content,"html.parser")
# print(new_response.content)

    dic["title"].append(get_title(new_soup))
    dic["price"].append(get_price(new_soup))
    dic["rating"].append(get_rating(new_soup))
    dic["review"].append(get_review(new_soup))

df=pd.DataFrame.from_dict(dic)
df['title'].replace('',np.nan,inplace=True)
df=df.dropna(subset=['title'])
df.to_csv('amazon_data.csv',header=True)
```
